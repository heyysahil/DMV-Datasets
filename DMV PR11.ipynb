### __Task 1:  Import the "City_Air_Quality.csv" dataset.__

```python
data = pd.read_csv(r"C:\Users\Student\Downloads\data.csv", encoding="cp1252") 
data
```

### __Task 2: Explore the dataset to understand its structure and content.__

```python
data.info()
```

```python
# Cleaning up name changes
data.state = data.state.replace({'Uttaranchal':'Uttarakhand'}) 
data.state[data.location == "Jamshedpur"] = data.state[data.location == 'Jamshedpur'].replace({"Bihar":"Jharkhand"})
```

```python
# Changing types to uniform format
types = {
    "Residential": "R", 
    "Residential and others": "RO",
    "Residential, Rural and other Areas": "RRO", 
    "Industrial Area": "I",
    "Industrial Areas": "I", 
    "Industrial": "I", 
    "Sensitive Area": "S", 
    "Sensitive Areas": "S", 
    "Sensitive": "S", 
    np.nan: "RRO"
}

data.type = data.type.replace(types) 
data.head()
```

### __Task 3:  Identify the relevant variables for visualizing AQI trends, such as date, pollutant levels, and AQI values.__

```python
# defining columns of importance, which shall be used reguarly 
VALUE_COLS = ['so2', 'no2', 'rspm', 'spm', 'pm2_5']
```

```python
# invoking SimpleImputer to fill missing values
imputer = SimpleImputer(missing_values=np.nan, strategy='mean') 
data[VALUE_COLS] = imputer.fit_transform(data[VALUE_COLS])
```

```python
# checking to see if the dataset has any null values left over and the format 
print(data.isnull().sum())
data.tail()
```

### __Task 4:  Create line plots or time series plots to visualize the overall AQI trend over time.__

```python
# Plotting highest and lowest ranking states
# defining a function to find and plot the top 10 and bottom 10 states for a given indicator (defaults to SO2) 
def top_and_bottom_10_states(indicator="so2"):
    fig, ax = plt.subplots(2, 1, figsize=(20, 12))
    
    # Calculate median values for each state
    ind = data[[indicator, 'state']].groupby('state', as_index=False).median().sort_values(by=indicator, ascending=False)
    
    # Top 10 states line plot
    top10_states = ind[:10]
    sns.lineplot(x='state', y=indicator, data=top10_states, ax=ax[0], marker='o', color='red')
    ax[0].set_title("Top 10 states by {} (1991-2016)".format(indicator))
    ax[0].set_ylabel("{} (Âµg/mÂ³)".format(indicator))
    ax[0].set_xlabel("State")
    ax[0].set_xticklabels(top10_states['state'], rotation=45)

    # Bottom 10 states line plot
    bottom10_states = ind[-10:]
    sns.lineplot(x='state', y=indicator, data=bottom10_states, ax=ax[1], marker='o', color='green')
    ax[1].set_title("Bottom 10 states by {} (1991-2016)".format(indicator))
    ax[1].set_ylabel("{} (Âµg/mÂ³)".format(indicator))
    ax[1].set_xlabel("State")
    ax[1].set_xticklabels(bottom10_states['state'], rotation=45)

    plt.tight_layout()
    plt.show()

# Call the function for different indicators
top_and_bottom_10_states("so2")
top_and_bottom_10_states("no2")
```

### __Task 5: Use bar plots or stacked bar plots to compare the AQI values across different dates or time periods.__

```python
# Plotting the highest ever recorded levels
# defining a function to find the highest ever recorded levels for a given indicator (defaults to SO2) by state 
# sidenote: mostly outliers
def highest_levels_recorded(indicator="so2"): 
    plt.figure(figsize=(20,10))
    ind = data[[indicator, 'location', 'state', 'date']].groupby('state', as_index=False).max() 
    highest = sns.barplot(x='state', y=indicator, data=ind)
    highest.set_title("Highest ever {} levels recorded by state".format(indicator))
    plt.xticks(rotation=90) 
    
highest_levels_recorded("no2") 
highest_levels_recorded("rspm")
```

### __Task 6: Create box plots or violin plots to analyze the distribution of AQI values for different pollutant categories.__

```python
def type_avg(indicator):
    plt.figure(figsize=(12, 6))
    
    # Create a box plot
    sns.boxplot(x='type', y=indicator, data=data)
    
    plt.title("Pollutant Distribution by Type for {}".format(indicator))
    plt.xlabel("Type")
    plt.ylabel("{} (Âµg/mÂ³)".format(indicator))
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()

# Call the function with the desired indicator
type_avg('so2')
```

### __Task 7: Use scatter plots or bubble charts to explore the relationship between AQI values and pollutant levels.__

```python
# Plotting pollutant average by type
# defining a function to plot pollutant averages by type for a given indicator 
def type_avg(indicator):
    # Assuming 'data' is your DataFrame
    # Calculate averages or other necessary values based on 'indicator'
    avg_data = data.groupby('type')[indicator].mean().reset_index()

    # Prepare sizes for bubbles (e.g., using counts or another metric)
    sizes = avg_data[indicator] * 10  # Adjust multiplier as needed

    plt.scatter(avg_data['type'], avg_data[indicator], s=sizes, alpha=0.5)
    plt.xticks(rotation=0)
    plt.title("Pollutant average by type for {}".format(indicator))
    plt.xlabel("Type")
    plt.ylabel(indicator)
    plt.show()

# Call the function with the desired indicator
type_avg('so2')
```

### __Task 8:  Customize the visualizations by adding labels, titles, legends, and appropriate color schemes__

```python
# Plotting pollutant averages by locations/state
# defining a function to plot pollutant averages for a given indicator (defaults to SO2) by locations in a given state 
def location_avgs(state, indicator="so2"):
    locs = data[VALUE_COLS + ['state', 'location', 'date']].groupby(['state', 'location']).mean() 
    state_avgs = locs.loc[state].reset_index()
    sns.barplot(x='location', y=indicator, data=state_avgs) 
    plt.title("Location-wise average for {} in {}".format(indicator, state)) 
    plt.xticks(rotation = 90)

location_avgs("Maharashtra", "no2")
```
