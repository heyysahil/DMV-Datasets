# __Title: Analyzing Sales Performance by Region in a Retail Company__

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### __Task 1: Import the "Retail_Sales_Data.csv" dataset.__

```python
df = pd.read_csv(r"C:\Users\Student\Downloads\customer_shopping_data.csv") 
df.head()
```

### __Task 2: Explore the dataset to understand its structure and content.__

```python
df.tail()
```

### __Task 3:  Identify the relevant variables for aggregating sales data, such as region, sales amount, and product category.__

```python
# To check the count of records grouped by region/branch of the mall
df.groupby("shopping_mall").count()
```

```python
# To check the count of records grouped by the product categories
df.groupby("category").count()
```

```python
# total sales for each mall branch
branch_sales = df.groupby("shopping_mall").sum()
branch_sales
```

```python
# total sales for each category of product
category_sales = df.groupby("category").sum()
category_sales
```

```python
# to get the top performing branches
branch_sales.sort_values(by = "price", ascending = False)
```

```python
# to get the top selling categories
category_sales.sort_values(by = "price", ascending = False)
```

### __Task 4:  Group the sales data by region and calculate the total sales amount for each region__

```python
# to get total sales for each combination of branch and product_category
combined_branch_category_sales = df.groupby(["shopping_mall", "category"]).sum()
combined_branch_category_sales
```

### __Task 5: Create bar plots or pie charts to visualize the sales distribution by region.__

```python
# pie chart for sales by branch
plt.pie(branch_sales["price"], labels = branch_sales.index) 
plt.show()
```

### __Task 6:  Identify the top-performing regions based on the highest sales amount.__

```python
# pie chart for sales by product category
plt.pie(category_sales["price"], labels = category_sales.index) 
plt.show()
```

### __Task 7:   Group the sales data by region and product category to calculate the total sales amount for each combination__

```python
combined_pivot = df.pivot_table(index="shopping_mall", columns="category", values="price", aggfunc="sum") 
```

### __Task 8: Create stacked bar plots or grouped bar plots to compare the sales amounts across different regions and product categories.__


```python
# grouped bar chart for sales of different categories at different branches
combined_pivot.plot(kind="bar", figsize=(10, 6)) 
plt.show()
```
